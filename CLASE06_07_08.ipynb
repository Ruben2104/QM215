```python
def generaDias(nEstudiantes):
  dias = np.random.randint(1, 366,size=nEstudiantes)
  op=False
  for i in range(0,nEstudiantes):
    n1=dias[i]
    v=dias[i+1:]
    for j in range(0,len(v)):
      if n1==v[j]:
        op=True
  return op
```

```python
n=0
for i in range(1000):
  if generaDias(20)==True:
    n=n+1
print(n/1000*100)
```

Reacci√≥n Quimica A--> P    K
 image.png

```python
A=1.0
p=0.0
K=0.1
t=0
dt=0.1
tt=[]
AA=[]
pp=[]
while t<50:
  tt.append(t)
  AA.append(A)
  pp.append(p)
  dA=-K*A*dt
  dp=K*A*dt
  A=A+dA
  p=p+dp
  t+=dt
tt=np.array(tt)
AA=np.array(AA)
pp=np.array(pp)

x=np.arange(0,50,5)
y=np.exp(-K*x)
y2=1-y
plt.plot(tt,AA)
plt.plot(tt,pp)
plt.scatter(x,y,color='black',s=8)
plt.scatter(x,y2,color='black',s=8)
plt.xlabel("t(seg)")
plt.ylabel("[A]")
plt.title("Reacci√≥n de primer orden A-->P")
plt.grid(True)
plt.show()
```

A+A -->  P

image.png

```python
A=1.0
p=0.0
K=0.1
t=0
dt=0.1
tt=[]
AA=[]
pp=[]
while t<50:
  tt.append(t)
  AA.append(A)
  pp.append(p)
  dA=-K*A*A*dt
  dp=K*A*A*dt
  A=A+dA
  p=p+dp
  t+=dt
tt=np.array(tt)
AA=np.array(AA)
pp=np.array(pp)

x=np.arange(0,50,2)
Ao=1.0
invAt=K*x+1/Ao
y=1/invAt

plt.plot(tt,AA)
plt.plot(tt,pp)
plt.scatter(x,y,color="black",s=8)
plt.xlabel("t(seg)")
plt.ylabel("[A]")
plt.title("Reacci√≥n de segundo orden A+A-->P")
plt.grid(True)
plt.show()
```

```python
A --> X    k1    vA=-K1*[A]    vX=K1*[A]-K2*[X]     vP=K2*[x]
X --> P    K2
```

```python
A=1.0
p=0.0
X=0.0
K1=0.1
K2=0.2
t=0
dt=0.1
tt=[]
AA=[]
pp=[]
XX=[]
while t<50:
  tt.append(t)
  AA.append(A)
  pp.append(p)
  XX.append(X)
  dA=-K1*A*dt
  dX=K1*A*dt-K2*X*dt
  dp=K2*X*dt
  A=A+dA
  X=X+dX
  p=p+dp
  t+=dt
tt=np.array(tt)
AA=np.array(AA)
pp=np.array(pp)
XX=np.array(XX)



plt.plot(tt,AA)
plt.plot(tt,XX)
plt.plot(tt,pp)

plt.xlabel("t(seg)")
plt.ylabel("[conc]")
plt.title("Reacci√≥n consecutiva A-->X  X-->P")
plt.grid(True)
plt.show()
```
